```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
%matplotlib inline
import pandas_datareader as web

ticker=["NKE"]

price_data = web.get_data_yahoo(ticker, start = '2000-1-3', end  = '2000-6-1')
price_data['MA10'] = price_data['Close'].rolling(10).mean()
price_data['MA50'] = price_data['Close'].rolling(50).mean()
price_data = price_data.dropna()
price_data.head()
```

```python
price_data['Shares'] =  [1 if price_data.loc[ei, 'High'] > price_data.loc[ei, 'Low'] else 0 for ei in price_data.index]
```
